# Imports

```python
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
import lightgbm as lgb
```

# Train

```python
train_set = pd.read_csv('train/train_encoded_sin_location.csv')
```

```python
X, y = train_set.iloc[:,:-1],train_set.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Grid Search

```python
def binary_preds(preds):
    len_preds = len(preds)
    for i in range(len_preds):
        if preds[i] >= 0.5:
            preds[i] = 1
        else:
            preds[i] = 0
```

```python
learning_rate = [0.1, 0.3, 0.01, 0.001]
num_leaves = [10, 20, 40, 70]
max_depth = [20, 50, 70, 100]
metric = 'binary_logloss'
sub_features = [0.4, 0.5, 0.6]
```

```python
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
best_acc = 0
best_params = {}; params = {}
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
    
for lr in learning_rate:
    params['learning_rate'] = lr
    for sf in sub_features:
        params['sub_feature'] = sf
        for nl in num_leaves:
            params['num_leaves'] = nl
            for md in max_depth:
                params['max_depth'] = md
        
                gbm = lgb.train(params, lgb_train, num_boost_round=100, valid_sets=lgb_eval)
                preds = gbm.predict(X_test, num_iterarion=gbm.best_iteration)
                binary_preds(preds)
                acc = np.sqrt(accuracy_score(y_test, preds))
                if acc > best_acc:
                    best_params = params
                    best_acc = acc
        
print(best_params, best_acc)
```

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)

params = {'objective': 'binary', 'metric': 'binary_logloss', 'num_leaves': 40,\
          'learning_rate': 0.1, 'sub_feature': 0.6, 'max_depth': 100}

gbm = lgb.train(params, lgb_train, num_boost_round=100, valid_sets=lgb_eval)
preds = gbm.predict(X_test, num_iterarion=gbm.best_iteration)

binary_preds(preds)

acc = np.sqrt(accuracy_score(y_test, preds))
print("Accuracy Score: %f" % (acc))
```
