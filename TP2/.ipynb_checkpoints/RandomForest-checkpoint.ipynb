```python
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
train_set = pd.read_csv('train.csv')
train_set.head(10)
```

```python
pos_col_text = train_set.columns.get_loc('text')+1
train_set.insert(loc=pos_col_text, column='len_text', value=train_set['text'].iloc[:].str.len())
train_set.head(10)
```

### Enconding de columnas categóricas/strings

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk #pip install nltk
nltk.download('punkt')
```

```python
tf_idf_vec = TfidfVectorizer(use_idf=True, smooth_idf=False, ngram_range=(1,1), stop_words='english', tokenizer=nltk.word_tokenize)
tf_idf_data = tf_idf_vec.fit_transform(train_set['text'])
```

```python
col_names = tf_idf_vec.get_feature_names()
```

```python
tf_idf_vector = tf_idf_data.toarray()
```

```python
tf_idf_df = pd.DataFrame(tf_idf_vector, columns=col_names)
train_set_aux = train_set.join(other=tf_idf_df, how='inner', lsuffix="_")
```

```python
train_set_aux.head(10)
```

### Random Forest

```python
from sklearn.model_selection import train_test_split

col_names.append('id_')
col_names.append('len_text')
col_names.append('target_')

new_aux = train_set_aux[col_names]

new_aux.head(10)
```

```python
X, y = new_aux.iloc[:,:-1],new_aux.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)
preds = rf_model.predict(X_test)
acc = np.sqrt(accuracy_score(y_test, preds))
print("Accuracy Score: %f" % (acc))
```
